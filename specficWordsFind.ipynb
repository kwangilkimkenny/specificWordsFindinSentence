```python
#txt 파일을 불러온다.
text_input=[]

file=open("HarvardEssays.txt", 'r')

while (1):
    line=file.readline()

    try:escape=line.index('\n')
    except:escape=len(line)
    
    if line:
        text_input.append(line[0:escape])
    else:
        break
    
file.close()
```

```python
print(text_input)
```

```python
# 문장에서 특정 단어 찾기
search = "KSA"

if search in str(text_input):
    print("Matching : {}".format(search))
   
else:
    print("Not Match")
```

```python
#문자를 찾아내고 위치반환
key_words = 'KSA'
lenth_setence = len(str(text_input)) #입력문장 전체길이
print("입력문장의 전체길이 : {} = 100%".format(lenth_setence))


pos_a = (str(text_input).find(key_words) * 100)/lenth_setence #문자의 상대적 위치 계산
a= int(pos_a) #정수로 변환
print("{}/100의 위치에 첫번째 키워드가 발견됨".format(a))
print("{0}를 100이라고 하면 {1}/100의 위치에 특정 단어가 위치하고 있음".format(lenth_setence, a))

print("전체문장에서 특정 키워드의 위치는 다음과 같음")

result_pos = []
while str(text_input)[a+1:].find(key_words) != -1:
    a = str(text_input)[a+1:].find(key_words) + a + 1
    pos_a = (a *100)/lenth_setence
    print("{}/100".format(int(pos_a)))
    result_pos.append(pos_a)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = result_pos
y = range(len(result_pos))

plt.plot(x, y, 'or')
plt.xlabel('Relative position')
plt.ylabel('keywords')
plt.title('The position of the keyword in the input sentence ')
plt.show()
```
